# Cleaning

**NULL VALUES**

```python
# get numeric columns
lsdf_num = lsdf.select_dtypes(include=[np.number])

lsdf_num
```

```python
type(lsdf_num.columns)
```

```python
for col in lsdf_num.columns:
    print(col + ": ", lsdf_num[col].isnull().values.any())
```

**DATA TYPE CONVERSIONS**

Object -> Dates:

```python
lsdf['Date'] = pd.to_datetime(lsdf['Date'])
```

```python
lsdf.Date.head(15)
```

Currency -> Numeric:

```python
lsdf.head()
```

```python
#Sale (Dollars)
lsdf['Sale (Dollars)'] = main_df['Sale (Dollars)'].astype(str)
lsdf['Sale (Dollars)'] = lsdf['Sale (Dollars)'].map(lambda x: x.replace('$',''))
lsdf['Sale (Dollars)'] = lsdf['Sale (Dollars)'].astype(str).astype(float)
```

```python
#State Bottle Retail
lsdf['State Bottle Retail'] = main_df['State Bottle Retail'].astype(str)
lsdf['State Bottle Retail'] = lsdf['State Bottle Retail'].map(lambda x: x.replace('$',''))
lsdf['State Bottle Retail'] = lsdf['State Bottle Retail'].astype(str).astype(float)
```

```python
#State Bottle Cost
lsdf['State Bottle Cost'] = main_df['State Bottle Cost'].astype(str)
lsdf['State Bottle Cost'] = lsdf['State Bottle Cost'].map(lambda x: x.replace('$',''))
lsdf['State Bottle Cost'] = lsdf['State Bottle Cost'].astype(str).astype(float)
```

```python
lsdf.head()
```

Object -> String

```python
# Store Name  
lsdf['Store Name'] = lsdf['Store Name'].astype(str)                     
# Address     
lsdf['Address'] = lsdf['Address'].astype(str)                     
# City    
lsdf['City'] = lsdf['City'].astype(str) 
# Store Location 
lsdf['Store Location'] = lsdf['Store Location'].astype(str)                     
# County   
lsdf['County'] = lsdf['County'].astype(str)   
# Category Name   
lsdf['Category Name'] = lsdf['Category Name'].astype(str)
# Vendor Name  
lsdf['Vendor Name'] = lsdf['Vendor Name'].astype(str)     
# Item Description    
lsdf['Item Description'] = lsdf['Item Description'].astype(str) 
```
